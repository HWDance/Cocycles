```python
trial = 10
n = len(results[0]['counterfactual_diffs'])
KSD = torch.zeros((trial,4))
KSD_split = torch.zeros((trial,4))
KSD_int = torch.zeros((trial,4))
KSD_int_split = torch.zeros((trial,4))
counterfactual_diffs = torch.zeros((trial,4))
counterfactual_diffs_split = torch.zeros((trial,4))
indexes = torch.zeros((trial,4))
indexes_split = torch.zeros((trial,4))
t = 0
for i in range(trial):
    for j in range(4):
        KSD[i,j] = results[t]['KSD']
        KSD_split[i,j] = results[t]['KSDsplit']
        KSD_int[i,j] = results[t]['KSDint']
        KSD_int_split[i,j] = results[t]['KSDintsplit']
        counterfactual_diffs[i,j] = ((results[t]['counterfactual_diffs'] - results[t]['counterfactual_diffpreds'])**2).mean()**0.5
        counterfactual_diffs_split[i,j] = ((results[t]['counterfactual_diffs'] - results[t]['counterfactual_diffpreds_split'])**2).mean()**0.5
        indexes[i,j] = results[t]['model_index']
        indexes_split[i,j] = results[t]['model_index_split'][0]
        t += 1
```
