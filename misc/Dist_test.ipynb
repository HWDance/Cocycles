```python
import torch
from torch.distributions import Gamma, MixtureSameFamily, Categorical, Normal, Laplace
import matplotlib.pyplot as plt
```

```python
x = 1/Gamma(2,1).sample((10000,))
plt.hist(x, bins = 100);
```

```python
n_mix = 10
mix = Categorical(1/torch.linspace(1,n_mix,n_mix))
comp = Laplace(2**torch.linspace(1,n_mix,n_mix), 2**torch.linspace(1,n_mix,n_mix)/2**0.5)
gmm = MixtureSameFamily(mix, comp)
x = gmm.sample((100000,))
plt.hist(x, bins = 250);
```
